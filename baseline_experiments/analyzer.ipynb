```python
import pandas as pd
import numpy as np

df = pd.read_csv('/home/ihossain/ISMAIL/SSL-malware/baseline_experiments/fixmatch/results/fixmatch_hidistloss_w_al_random_bit_flip_4_lbr_0.4_seed_5_ssl_loss.csv')
```

```python
import matplotlib.pyplot as plt
# === PLOT 1: Total, Supervised, and Unsupervised Loss ===
plt.figure(figsize=(10, 6))
plt.plot(df['epoch'], df['loss'], label='Total Loss')
plt.plot(df['epoch'], df['sup_loss'], label='Supervised Loss')
plt.plot(df['epoch'], df['unsup_loss'], label='Unsupervised Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss Components Over Epochs')
plt.legend()
plt.grid(True)
plt.tight_layout()
# plt.savefig(save_plot1)
plt.show()
# print(f"Saved Plot 1 to: {save_plot1}")
```

```python
# === PLOT 2: Supervised, Contrastive (Lhc), and Classification (Lce) Loss ===
plt.figure(figsize=(10, 6))
plt.plot(df['epoch'], df['sup_loss'], label='Supervised Loss')
plt.plot(df['epoch'], df['lhc_val'], label='Contrastive Loss (Lhc)')
plt.plot(df['epoch'], df['lce_val'], label='Classification Loss (Lce)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Supervised, Contrastive and Classification Loss Over Epochs')
plt.legend()
plt.grid(True)
plt.tight_layout()
# plt.savefig(save_plot2)
plt.show()
# print(f"Saved Plot 2 to: {save_plot2}")
```
